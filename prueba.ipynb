```
# llamar las bases
tiendas = pd.read_csv("/content/Base_Tiendas.csv")
ventas = pd.read_csv("/content/Base_Ventas.csv")
jerarquias = pd.read_csv("/content/JERARQUIAS.csv")
```

```
# Renombrar columnas en df_ventas para unir por STORE y jerarquía
ventas.rename(columns={
    '/BIC/ZCLCH_SMA': 'STORE',
    '/BIC/ZCLCH_LIN': 'CATEGORIA',
    '/BIC/ZCLCH_CHA': 'JERARQUIA2'
}, inplace=True)
```

```
ventas
```

```
for col in ventas.columns:
    print(repr(col))
```

```
ventas.columns
```

```
ventas_tiendas["JERARQUIA2"] = ventas_tiendas["JERARQUIA2"].astype(str)
jerarquias["JERARQUIA2"] = jerarquias["JERARQUIA2"].astype(str)

```

```
ventas_full = pd.merge(ventas_tiendas, jerarquias, how='left', on='JERARQUIA2')
```

```
ventas_full[['JERARQUIA2', 'NOMBRE JERARQUIA2']].head()
```

```
ventas_full['NOMBRE JERARQUIA2'].isnull().sum()
```

```
ventas_resumen = ventas_full.groupby([
    'NOMBRE JERARQUIA2', 'STORENAME', 'CATEGORIA'
]).agg({
    'VENTA_BRUTA': 'sum',
    'IVA': 'sum',
    'UNIDADES': 'sum'
}).reset_index()
```

```
ventas_resumen.to_csv("ventas_por_jerarquia_tienda_categoria.csv", index=False)

from google.colab import files
files.download("ventas_por_jerarquia_tienda_categoria.csv")
```
